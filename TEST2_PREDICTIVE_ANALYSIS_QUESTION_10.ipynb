```python
import numpy as np
import pandas as pd
from scipy.stats import kendalltau
from scipy.stats import pearsonr

import sys
sys.path.append('../')

def solution():
    # Write code here 
    df = pd.read_csv('res/autompg.csv')

    #question1 output/output1.csv
    df_mean = df['mpg'].mean()
    df_mean_mpg = str(round(df_mean,2))
    df_mean_mpg
    df_mean_mpg.to_csv('output/output1.csv')
    print('df_mean_mpg',df_mean_mpg)
    
     
    

    df_median = df['mpg'].median()
    df_median_mpg = str(round(df_median,2))
    df_median_mpg
    df_median_mpg.to_csv('output/output2.csv')
    print('df_median_mpg',df_median_mpg)
    
    

    df_mode = df['mpg'].mode()
    df_mode_mpg = str(round(df_mode,2))
    df_mode_mpg
    df_mode_mpg.to_csv('output/output3.csv')
    print('df_mode_mpg',df_mode_mpg)
    
    

    df_std = df['mpg'].std()
    df_std_mpg = str(round(df_std,2))
    df_std_mpg
    df_std_mpg.to_csv('output/output4.csv')
    print('df_std_mpg',df_std_mpg)
    

    #question 2
    correlation = df.corr()
    pearson_coef = correlation.loc['weight','mpg']
    kendalltau_coef = correlation.loc['weight','mpg']
    print('kendalltau_coef.to_csv("output/output5.csv")',kendalltau_coef.mean().round(2))
    print('pearson_coef.to_csv("output/output6.csv")',pearson_coef.mean().round(2))

    #question 3

    df1 = pd.read_csv('res/autosample1.csv')

    df1_mean = df1['mpg'].mean()
    df1_mean_mpg = str(round(df1_mean,2))
    df1_mean_mpg
    df1_mean_mpg.to_csv('output/output7.csv')
    print('df1_mean_mpg',df1_mean_mpg)

    df1_std = df1['mpg'].std()
    df1_std_mpg = str(round(df1_std,2))
    df1_std_mpg
    df1_std_mpg.to_csv('output/output8.csv')
    print('df1_std_mpg',df1_std_mpg)


    #step3
    difference = df_mean_mpg- df1_mean_mpg
    output = str(round(difference,2))
    output.to_csv('output/output9.csv')
    print('difference of population and sample',output)
```
